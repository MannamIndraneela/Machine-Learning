### Draw the different Bivariate plots for All the Numerical variables in the Dataset

```python
df[df['model year'] < 80].sample(100).plot.scatter(x='model year',y='acceleration')
```

```python
df[df['origin'] < 100].plot.hexbin(x='mpg', y='displacement', gridsize=15)
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.area()
```

```python
df.plot.line()
```

### Draw the Bivariate plots for combination of Numerical and Categorical variables in the Dataset

```python
df_origin = df.groupby(['origin', 'car name']).mean()[['mpg', 'cylinders']]
df_origin.plot.bar(stacked=True)
```

```python
df_stats_by_displacement = df.groupby('displacement').mean()[['mpg', 'cylinders', 'weight', 'acceleration', 'model year', 'origin']]
```

```python
df_stats_by_displacement.plot.line()
```
